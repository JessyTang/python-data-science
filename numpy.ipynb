**创建数组**
---
1. np.array([list])
    可以通过list生成相应ndarray
2. np.zeros(10, dtype=int)
    创建一个长度为10的数组，值均为0
3. np.ones((3, 5), dtype=int)
    创建一个3x5的二维数组，值均为1
4. np.full((3, 5), 3.14)
    创建一个3x5的二维数组，值均为3.14
5. np.arange(0, 20, 2)
    创建从0开始到20（不包含）的数组，步长为2
6. np.linspace(0, 1, 5)
    创建从0开始到1的等间距数组，支持5个索引Index
7. np.random.random(0, 1, (3, 5))
    创建3x5的二维数组，值服从N（0, 1）正态分布
8. np.random.randint(0, 11, (3, 5))
    创建3x5的二维数组，值是在[0, 10]之间的整数
9. np.eye(3)
    创建3阶单位阵I
10. np.empty((3, 5))
    创建3x5的二维数组，对值不进行初始化

```python
np.array([1, 2, 3, 4, 5])
np.zeros(10, dtype=int)
np.ones((3, 5), dtype=int)
np.full((3, 5), 3.14)
np.arange(0, 20, 2)
np.linspace(0, 1, 5)
np.random.random((3, 5))
np.random.normal(0, 1, (3, 5))
np.random.randint(0, 11, (3, 5))
np.eye(3)
np.empty((3, 5), dtype=int)
```
